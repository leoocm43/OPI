```
#Creando DataFrame a partir del archivo demo.csv
demo = spark.read.format('csv').options(header='true', inferSchema='true').load('/FileStore/tables/demo.csv')
#demo2 = demo.filter("country is not null")

#Se toma el DataFrame demo para convertirlo a diccionario
keys = demo.select("country").collect()
values = demo.select(demo["name"],demo["alias"],demo["iata"],demo["icao"],demo["callsign"],demo["active"],demo["airline_id"]).collect()
dic = dict.fromkeys(keys,values)

# Se escribe el diccionario en un archivo csv
dicdemo = spark.createDataFrame(dic)
writedemo = dicdemo.write.format('csv').options(header = "true").save('/FileStore/tables/writedemo.csv')

```

```
#Se crea un DataFrame a partir del archivo csv creado anteriormente tomando como fuente el diccionario dic.
democompare = spark.read.format('csv').options(header='true', inferSchema='true').load('/FileStore/tables/writedemo.csv')

#Se hace el inner join
newdemo = demo.join(democompare,"country")
```
